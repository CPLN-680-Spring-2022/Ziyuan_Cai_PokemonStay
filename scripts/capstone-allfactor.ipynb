```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
def read_factor(file_path):
    data = gpd.read_file(file_path)
    data = data[["geometry"]]
    return data
```

```python
def count_factor(data,column_name):
    data500 = gpd.sjoin(data,des,how='inner',op='intersects')
    data500_group = data500.groupby(['FID'], as_index=False)["index_right"].count()
    data500_group = des.merge(data500_group,on="FID",how="left")
    data500_group = data500_group.fillna(0)
    data500_group = data500_group.rename(columns={"index_right": column_name})
    return data500_group
```

```python
def contain_factor(data,column_name):
    data500 = gpd.sjoin(data,des,how='inner',op='intersects')
    data500_group = data500.groupby(['FID'], as_index=False)["index_right"].count()
    data500_group["index_right"] = 1
    data500_group = des.merge(data500_group,on="FID",how="left")
    data500_group = data500_group.fillna(0)
    data500_group = data500_group.rename(columns={"index_right": column_name})
    return data500_group
```

```python
#read destination and set 500 meter buffer zone
des=gpd.read_file("E:/graduate/capstone/capstone/processed_data/destination.shp")
des = des.to_crs(epsg=3857)
des.geometry=des.geometry.buffer(500)
des = des.to_crs(epsg=4326)
```

```python
#crime
crime = read_factor("E:/graduate/capstone/capstone/processed_data/crime.shp")
crime500_group = count_factor(crime,"total_crime")
```

```python
fig, ax = plt.subplots(figsize=(12, 12))

# Plot
crime500_group.plot(
    ax=ax, 
    column="total_crime", 
    edgecolor="white", 
    linewidth=0.5, 
    legend=True,
    legend_kwds=dict(loc="lower right"),
    cmap="viridis",
    scheme="Quantiles",
    k=5
)
```
