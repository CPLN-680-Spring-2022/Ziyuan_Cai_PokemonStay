```python
# Select Frankford Hall based on the name column
frankford_hall = food.query("name == 'Powelton Pizza'").squeeze()

# Get the x/y coordinates
frankford_hall_x = frankford_hall.geometry.x
frankford_hall_y = frankford_hall.geometry.y
```

```python
# Select Spring Garden based on the name column
spring_garden_station = busstop.query("name == '13th St & Market St'").squeeze()

# Get the x/y coordinates
spring_garden_x = spring_garden_station.geometry.x
spring_garden_y = spring_garden_station.geometry.y
```

```python
# Get the origin node
orig_node = ox.distance.nearest_nodes(philly,  spring_garden_x, spring_garden_y) 

# Get the destination node
dest_node = ox.distance.nearest_nodes(philly, frankford_hall_x, frankford_hall_y) 
```

```python
# get the shortest path --> just a list of node IDs
route = nx.shortest_path(philly, 
                         orig_node, 
                         dest_node, 
                         weight='length')
route
```

```python
ox.plot_graph_route(philly, route, node_size=0);
```
