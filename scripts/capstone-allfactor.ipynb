```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
def read_factor(file_path):
    data = gpd.read_file(file_path)
    data = data[["geometry"]]
    return data
```

```python
def count_factor(data,column_name):
    data500 = gpd.sjoin(data,des,how='inner',op='intersects')
    data500_group = data500.groupby(['FID'], as_index=False)["index_right"].count()
    data500_group = des.merge(data500_group,on="FID",how="left")
    data500_group = data500_group.fillna(0)
    data500_group = data500_group.rename(columns={"index_right": column_name})
    return data500_group
```

```python
def contain_factor(data,column_name):
    data500 = gpd.sjoin(data,des,how='inner',op='intersects')
    data500_group = data500.groupby(['FID'], as_index=False)["index_right"].count()
    data500_group["index_right"] = 1
    data500_group = des.merge(data500_group,on="FID",how="left")
    data500_group = data500_group.fillna(0)
    data500_group = data500_group.rename(columns={"index_right": column_name})
    return data500_group
```

```python
crs = 4326
city_limits = gpd.read_file('E:/graduate/capstone/capstone/raw_data/City_Limits.geojson')
city_limits = city_limits.to_crs(crs)
```

```python
x_min, y_min, x_max, y_max = city_limits.total_bounds

#find start point
while 1:
    x = np.random.uniform(x_min, x_max,2)
    y = np.random.uniform(y_min, y_max,2)

    # convert it to a points GeoSeries
    start = gpd.GeoSeries(gpd.points_from_xy(x, y,crs=4326))
    # only keep those points within polygons
    start = start[start.within(city_limits)]
    start = gpd.GeoDataFrame(start, crs=4326, geometry=start)
    if len(start)==1:
        break
```

```python
distance = des.geometry.apply(lambda g: start.distance(g))
```

```python
distance = pd.concat([des_origin,distance],axis=1)
distance = distance.rename(columns={0: "distance"})
```

```python
distance["distance"]=(max(distance["distance"])-distance["distance"])/(max(distance["distance"])-min(distance["distance"]))
distance
```

```python
def df_to_geojson(df, properties):
    # create a new python dict to contain our geojson data, using geojson format
    geojson = {'type':'FeatureCollection', 'features':[]}

    # loop through each row in the dataframe and convert each row to geojson format
    for _, row in df.iterrows():
        # create a feature template to fill in
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}

        # fill in the coordinates
        feature['geometry']['coordinates'] = [row.geometry.x,row.geometry.y]

        # for each column, get the value and add it as a new feature property
        for prop in properties:
            feature['properties'][prop] = row[prop]
        
        # add this feature (aka, converted dataframe row) to the list of features inside our dict
        geojson['features'].append(feature)
    
    return geojson
```

```python
cols = ['total_pokestop', 'total_crime', 'contain_busstop', 'contain_drink', 'contain_food', 'contain_gym', 'contain_nest', 'contain_raid']
geojson = df_to_geojson(factor, cols)
```
